```python
%matplotlib inline
import numpy as np
np.random.seed(42)
import seaborn as sns
sns.set()
```

```python
cost_savings_mean = 5
cost_savings_std  = 4
market_size_mean  = 40000
market_size_std   = 10000
num_simulations   = 1000
```

```python
savings = np.random.normal(cost_savings_mean, cost_savings_std, num_simulations) * np.random.normal(market_size_mean, market_size_std, num_simulations)
```

```python
sns.distplot(savings)
```

```python
f"Average savings of {np.mean(savings)}"
```
