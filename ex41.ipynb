```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Markdown as md
np.random.seed(42)

%matplotlib inline
```

Considering the regression: $log(weight) = -3.5 + 2.0*log(height) + \epsilon$ with errors having $\sigma = 0.25$

Ex 4.1.a: Approximately $68%$ of the persons will have weights within a factor of `__` and `__` of their predicted values from the regression.

```python
esd = np.exp(0.25)
md(f"68% -> 1 std -> log transformation -> -exp(0.25) to exp(0.25) -> -{esd} to {esd}")
```

```python
heights_reg = np.arange(140, 210, 1)
heights = np.random.normal(175, 10, 200) # pretty random here
werr = np.random.normal(0, 0.25, heights.shape)
log_weights = -3.5 + 2*np.log(heights) + werr
log_weights_reg = -3.5 + 2*np.log(heights_reg)
```

```python
ax = sns.scatterplot(heights, log_weights);
sns.lineplot(heights_reg, log_weights_reg, ax=ax);
```

```python
ax = sns.scatterplot(heights, np.exp(log_weights));
sns.lineplot(heights_reg, np.exp(log_weights_reg), ax=ax);
```
