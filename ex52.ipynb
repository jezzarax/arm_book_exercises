```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
xd = np.arange(-10, 10, 0.1)
```

```python
inv_logit = lambda x: 1/(1 + np.exp(-x))

ya = inv_logit(xd)
yb = inv_logit(2 + xd)
yc = inv_logit(2 * xd)
yd = inv_logit(2 + 2 * xd)
ye = inv_logit(-2 * xd)

df = pd.DataFrame({
    'x': xd,
    '(x)': ya,
    '(2+x)': yb,
    '(2x)': yc,
    '(2+2x)': yd,
    '(-2x)': ye,
}).melt(id_vars=['x'], var_name='fn', value_name='y')
```

```python
sns.lineplot(data=df, x='x', y='y', hue='fn')
```
