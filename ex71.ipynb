```python
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set()
np.random.seed(42)
```

```python
p_shot = 0.6
```

```python
def call_shots():
    shots = []
    while True:
        shots.append(np.random.binomial(1, p_shot, 1)[0] == 1)
        if len(shots) > 1:
            if shots[-1] and shots[-2]:
                break
    return shots
    
```

```python
num_simulations = 1000
shots = [call_shots() for _ in range(num_simulations)]
shots_to_success = np.array([len(s) for s in shots])
successful_shots = np.array([sum(s) for s in shots])
```

```python
sns.distplot(shots_to_success, kde=True)
```

```python
f"It takes on average {np.mean(shots_to_success)} shots to get to two consecutive successes, with standard deviation of {np.std(shots_to_success)} and median of {np.median(shots_to_success)}"
```

```python
sns.scatterplot(x=shots_to_success, y=successful_shots/shots_to_success)
```
