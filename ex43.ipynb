```python
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
sns.set()
```

```python
age = np.arange(0, 100, 1)
age10 = age/10.
weight_ex1 = age10*2.6 + 161
```

```python
ax = sns.lineplot(x = age, y = weight_ex1, markers=False);
ax.set_ylim(0, 200);
```

```python
age10_sq = np.power(age/10, 2)
weight_ex2 = 96.2 + age10*33.6 - age10_sq*3.2
```

```python
ax = sns.lineplot(x = age, y = weight_ex1, markers=False);
sns.lineplot(x = age, y = weight_ex2, markers=False, ax=ax);
ax.set_ylim(0, 200);
```
