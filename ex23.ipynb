```python
import numpy as np
import seaborn as sns
from scipy.stats import norm, normaltest
```

```python
samp_data = np.sum(np.random.uniform(size=(20, 1000)), axis=0)
sns.distplot(samp_data, fit=norm)
```

```python
normaltest(samp_data)
```

```python
samp_data = np.sum(np.random.uniform(size=(20, 100000)), axis=0)
sns.distplot(samp_data, fit=norm)
```

```python
normaltest(samp_data)
```
