The model given is $ Pr(y=1) = logit^{-1}(1+2x_1+3x_2) $

Given a data point $ x_1=1, x_2=0.5$

Estimating the $ Pr(y=1) $ for a given data point:

$ Pr(y=1) = Pr(1+2x_1+3x_2 > 0) = logit^{-1}(1+2x_1+3x_2) = logit^{-1}(1+2*1 + 3*0.5) = logit^{-1}(4.5) $

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Markdown as md

sns.set()
```

```python
inv_logit = lambda x: 1/(1 + np.exp(-x))
logit = lambda x: np.log(x) - np.log(1 - x)
```

```python
prob = inv_logit(4.5)
md(f"Probability of the data point having probability of 1 is {prob}")
```

```python
xd = np.arange(-10, 10, 0.1)
yd = inv_logit(xd)
ydd = np.diff(yd)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
sns.lineplot(x=xd[1:], y=ydd, ax=ax1)
ax1.fill_between(xd[1:], ydd, where=xd[1:]>4.5, facecolor='green', hatch=r"//");
ax1.set_ylabel('pdf')

sns.lineplot(x=xd, y=yd, ax=ax2)
ax2.fill_between(xd, yd, where=xd>4.5, facecolor='green', hatch=r"//");
ax2.set_ylabel('cdf')
fig.set_size_inches(16, 10)
```
