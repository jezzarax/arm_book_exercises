```python
import numpy as np
import seaborn as sns
from scipy.stats import norm
```

```python
mmean = 69.1
msd = 2.9

fmean = 63.7
fsd = 2.7

n_simulations = 1000
```

```python
mheights = np.random.normal(mmean, msd, size=(100, n_simulations))
fheights = np.random.normal(fmean, fsd, size=(100, n_simulations))
```

```python
h_diff = mheights - fheights
```

```python
h_diff.shape
```

```python
sns.distplot(np.average(h_diff, axis=0), fit=norm)
```
